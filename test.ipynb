```python
import pandas_datareader as pdr
from env import API_KEY
import pandas as pd
import plotly.express as px
```

```python
# Data Collection
from tiingo import TiingoClient
from datetime import datetime, timedelta

config = {
    'api_key': API_KEY,
    'session': True  # Use requests.Session
}
client = TiingoClient(config)

# Calculate the start date as today's date minus 2 years
end_date = datetime.now()
start_date = end_date - timedelta(days=20*365)

# Convert dates to string format
start_date_str = start_date.strftime('%Y-%m-%d')
end_date_str = end_date.strftime('%Y-%m-%d')

df = client.get_dataframe('AAPL', frequency='daily',startDate=start_date_str, endDate=end_date_str)
```

```python
df.head()
```

```python
df.reset_index(inplace=True)
df.rename(columns={'index': 'new_index'}, inplace=True)  # Rename the old index column
df['date'] = df['date'].dt.date  # Extract the date part from the datetime index
df
```

```python
# Convert the 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])
```

```python
#  extract month and year from dates
df['Month'] = [i.month for i in df['date']]
df['Year'] = [i.year for i in df['date']]
df['Day'] = [i.day for i in df['date']]
```

```python
import numpy as np 
# # create a sequence of numbers
# df['Series'] = np.arange(1,len(df)+1)
```

```python
df.drop(['volume','adjClose','adjHigh','adjLow','adjOpen','adjVolume','divCash',],axis=1,inplace=True)
df = df[['Year','Month','Day','close','high','low','open','date']]
df.head()
```

```python
df.head()
```

```python
# Drop the first row since it doesn't have a previous day's closing price to compare with
df = df.dropna()
```

```python
df_copy = df.copy()
df_copy
```

```python
df.drop(['date','high','low','open'],axis=1,inplace=True)
```

```python
df.head()
```

```python
# split data into train-test set
train = df[df['Year'] < 2023]
test = df[df['Year'] >= 2023]
# check shape
train.shape, test.shape
```

```python
# Setup the classification task with time series configuration
# import the regression module
from pycaret.regression import *
# initialize setup
close_s = setup(data=train, test_data=test, target='close', fold_strategy='timeseries', numeric_features=['Year'], data_split_shuffle=False, fold_shuffle=False, session_id=123)

```

```python
close_best = compare_models(sort = 'MAE')
```

# Percent Change in Stock Price

```python
import requests
import config as c
import json
```

```python
ticker = "TSLA"
url = "https://api.tiingo.com/tiingo/daily/"
headers = {
        'Content-Type': 'application/json'
        }
requestResponse = requests.get(f"{url}{ticker}/prices/?startDate=2012-1-1&endDate=2016-1-1&token={c.API_KEY}", headers=headers)
print(json.dumps(requestResponse.json(), indent=4))
```
